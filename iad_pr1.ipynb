```python
# MARINA
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans

 # --------  Загрузка датасета --------
N = np.random.randint(1, 1000)   # кількість кластерів
X1, y1 = make_blobs(n_samples=250, n_features=2, centers=N, random_state=0)
print(y1)
# -> make_blobs(n_samples=250, n_features=2, centers=8, random_state=0): Создает искусственный набор данных с 250 образцами, 2 признаками (или признаковыми переменными), распределенными по 8 центрам (кластеров).
# -> X1 содержит координаты данных (массив признаков), а y1 содержит метки кластеров, которые были сгенерированы для каждого образца.

 # --------  KMeans --------
kmeans = KMeans(n_clusters=8, random_state=0)
kmeans.fit(X1)
kmeans_pred_res = kmeans.predict(X1)
kmeans_ari = adjusted_rand_score(y1, kmeans_pred_res)
kmeans_centroinds = kmeans.cluster_centers_
print(f'Adjusted Rand Score for KMeans: {kmeans_ari}', '', sep='\n')
print('centroids', kmeans_centroinds, '', sep='\n')
print('prediction', kmeans_pred_res, sep='\n')
# -> kmeans = KMeans(n_clusters=8, random_state=0): Создает объект кластеризатора K-средних с 8 кластерами и фиксированным случайным состоянием для воспроизводимости результатов.
# -> kmeans.fit(X1): Обучает модель K-средних на данных X1.
# -> kmeans_pred_res = kmeans.predict(X1): Применяет обученную модель к данным X1, чтобы предсказать, к каким кластерам принадлежат образцы.
# -> kmeans_ari = adjusted_rand_score(y1, kmeans_pred_res): Вычисляет скоринг Рэнда для оценки качества кластеризации, сравнивая истинные метки кластеров y1 и предсказанные метки kmeans_pred_res.
# -> kmeans_centroinds = kmeans.cluster_centers_: Получает координаты центров кластеров, вычисленные алгоритмом K-средних.
# -> print(f'Adjusted Rand Score for KMeans: {kmeans_ari}', '', sep='\n'): Печатает скоринг Рэнда.
# -> print('centroids', kmeans_centroinds, '', sep='\n'): Печатает координаты центров кластеров.
# -> print('prediction', kmeans_pred_res, sep='\n'): Печатает предсказанные метки кластеров для каждого образца.

 # -------- Визуализация прогнозов --------------
# plt.subplot(1, 2, 1)
#plt.subplots_adjust(wspace=0.2)
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans_pred_res, cmap="rainbow")
plt.scatter(kmeans_centroinds[:, 0], kmeans_centroinds[:, 1], marker="x", color="black", s=100)
#plt.title("KMeans")
#plt.xlabel("Feature 1")
#plt.ylabel("Feature 2")
# -> plt.scatter(X1[:, 0], X1[:, 1], c=kmeans_pred_res, cmap="rainbow"): Создает рассеянный график, где точки окрашены в зависимости от их предсказанных меток кластеров. Цвета выбираются из цветовой карты "rainbow".
#-> plt.scatter(kmeans_centroinds[:, 0], kmeans_centroinds[:, 1], marker="x", color="black", s=100): Наносит центры кластеров на график в виде черных крестиков размером 100.
```

```python
# MARINA COPY WITH RANDOM
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.metrics import adjusted_rand_score
from sklearn.cluster import KMeans

# Определите количество образцов и признаков
n_samples = 250
n_features = 2

# Генерируем случайную матрицу данных X1
X1 = np.random.random((n_samples, n_features))

# Генерируем количество кластеров и метки кластеров
N = np.random.randint(1, min(20, n_samples))  # Ограничение на максимальное количество кластеров
y1 = np.random.randint(0, N, size=n_samples)  # Случайные метки кластеров от 0 до N-1
print("---- True cluster labels:", y1)

 # --------  KMeans --------
kmeans = KMeans(n_clusters=N, random_state=1)
kmeans.fit(X1)
kmeans_pred_res = kmeans.predict(X1)
kmeans_ari = adjusted_rand_score(y1, kmeans_pred_res)
kmeans_centroinds = kmeans.cluster_centers_
print(f'---- Adjusted Rand Score for KMeans: {kmeans_ari}', '', sep='\n')
print('---- centroids', kmeans_centroinds, '', sep='\n')
print('---- prediction', kmeans_pred_res, sep='\n')
# -> kmeans = KMeans(n_clusters=8, random_state=0): Создает объект кластеризатора K-средних с 8 кластерами и фиксированным случайным состоянием для воспроизводимости результатов.
# -> kmeans.fit(X1): Обучает модель K-средних на данных X1.
# -> kmeans_pred_res = kmeans.predict(X1): Применяет обученную модель к данным X1, чтобы предсказать, к каким кластерам принадлежат образцы.
# -> kmeans_ari = adjusted_rand_score(y1, kmeans_pred_res): Вычисляет скоринг Рэнда для оценки качества кластеризации, сравнивая истинные метки кластеров y1 и предсказанные метки kmeans_pred_res.
# -> kmeans_centroinds = kmeans.cluster_centers_: Получает координаты центров кластеров, вычисленные алгоритмом K-средних.
# -> print(f'Adjusted Rand Score for KMeans: {kmeans_ari}', '', sep='\n'): Печатает скоринг Рэнда.
# -> print('centroids', kmeans_centroinds, '', sep='\n'): Печатает координаты центров кластеров.
# -> print('prediction', kmeans_pred_res, sep='\n'): Печатает предсказанные метки кластеров для каждого образца.

 # -------- Визуализация прогнозов --------------
# plt.subplot(1, 2, 1)
#plt.subplots_adjust(wspace=0.2)
plt.scatter(X1[:, 0], X1[:, 1], c=kmeans_pred_res, cmap="rainbow")
plt.scatter(kmeans_centroinds[:, 0], kmeans_centroinds[:, 1], marker="x", color="black", s=100)

plt.title("KMeans")
# plt.xlabel("Feature 1")
# plt.ylabel("Feature 2")
plt.show()
# -> plt.scatter(X1[:, 0], X1[:, 1], c=kmeans_pred_res, cmap="rainbow"): Создает рассеянный график, где точки окрашены в зависимости от их предсказанных меток кластеров. Цвета выбираются из цветовой карты "rainbow".
#-> plt.scatter(kmeans_centroinds[:, 0], kmeans_centroinds[:, 1], marker="x", color="black", s=100): Наносит центры кластеров на график в виде черных крестиков размером 100.
```
